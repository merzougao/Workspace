# Integration

## Definitions

$L_1$ is the class of integrable functions

A non negative, measurable function $f$ is $\textbf{integrable}$ if $\int f < \infty$

A measurable function $f$ is $\textbf{integrable}$ if $f_+$ and $f_-$ are integrables 

The $\textbf{integral}$ of a function w.r.t the measure $\mu$ on the set $A$ is defined by :

\begin{align}
\int_A fd\mu = \mu(\mathbb{1}_Af)
\end{align}

## Integration of functions

- Integral of characteristic function 

> $(X,S,\mu)$ measure space, 

> $E\in S$

> $f = \mathbb{1}_E$

\begin{align}
\int fd\mu =  \mu(E)
\end{align}

- Integral of non negative, simple function

> $(X,S,\mu)$ measure space, 

>$E_1 \dots E_n$, disjoints sets of $S$ where $\bigcup E_i = X$

> $f = \sum_{i=1}^na_i\mathbb{1}_{E_i}$

\begin{align}
\int fd\mu =  \sum_{i=1}^na_i\mu(E_i)
\end{align}

- Integral of non negative, measurable function

> $(X,S,\mu)$ measure space, 

> $f$ non negative measurable function

> $(f_n)_{n \in \mathbb{N}} \uparrow f$ s.t $\forall i \;f_i$ simple function 

\begin{align}
\int fd\mu =  \lim_{n \to \infty}\int f_nd\mu
\end{align}

- Integral of non negative, measurable function a.e

> $(X,S,\mu)$ measure space, 

> $f$ non negative, measurable function a.e

> $g$ non negative, measurable function with $f = g$ a.e

\begin{align}
\int fd\mu =  \int gd\mu
\end{align}

- Integral of measurable function a.e

> $(X,S,\mu)$ measure space, 

> $f$ measurable function a.e

> $f_+$ and  $f_-$ non negative, measurable functions a.e whith $f = f_+ - f_-$

\begin{align}
\int f =  \int f_+ - \int f_-
\end{align}

## Theorems

- Monotone Convergence Theorem

> $(X,S,\mu)$ measure space

> $(f_n)_{n \in \mathbb{N}}$ monotone increasing sequence of non negative, measurable, functions a.e

> $f$ non negative, measurable function with $ \lim_{n \to \infty}f_n(x) = f(x)$, $x\in X$ a.e

\begin{align}
\lim_{n \to \infty}\int f_nd\mu = \int fd\mu
\end{align}

- Fatou's Lemma

> $(X,S,\mu)$ measure space

> $(f_n)_{n \in \mathbb{N}}$ monotone increasing sequence of non negative, measurable, functions a.e

\begin{align}
\lim_{n \to \infty}\inf\int f_nd\mu = \int(\lim_{n \to \infty}\inf f_n)d\mu
\end{align}

- Dominated convergence

> $(X,S,\mu)$ measure space

> $(f_n)_{n \in \mathbb{N}}$ sequence of integrable functions

> $g$ integrable function s.t $\forall n$, $\left|f_n\right| \leq \left|g\right|$ a.e

> f measurable function s.t $f_n(x) \rightarrow f(x)$ a.e

\begin{align}
f &\in L_1 \\ 
\int \left|f_n - f\right|d\mu &\rightarrow 0
\end{align}


- Transformation theorem

> $(X,S,\mu)$ measure space

> $(Y,\xi)$ measurable space

> $T : X \rightarrow Y$ Transformation defined a.e

> $f : Y \rightarrow \mathbb{R}$ measurable function

\begin{align}
\int_Y fd(\mu T^{-1}) = \int_X fTd\mu
\end{align}

- Radon-Nikodym theorem

> $(X,S,\mu)$ totally $\sigma$-finite measure space

> $ v << \mu $ with $v$ $\sigma$-finite signed measure on $S$

\begin{align}
&\exists \text{ f measurable } &f:X\rightarrow [0,\infty),\;  \\
&\forall E \in S &v(E) = \int fd\mu
\end{align}

Notationwise : $f = \cfrac{dv}{d\mu}$

- Lebesgue decomposition

> $(X,S)$ measurable space

> $\mu$ and $v$ $\sigma$ -finite signed measure on $S$

\begin{align}
\exists v_1, v_2 \;\; s.t \;\; v &= v_1 + v_2 \\
v_0 &\perp \mu \\
v_1 &<< \mu
\end{align}

- Generated measures

> $(X,S,\mu)$ measure space

> $g:X\rightarrow [0,\infty]$ measurable

> $v(E) = \int_E gd\mu$


\begin{align}
\text{Then $v$ is a measure on $(X, S)$ and for any $f:X\rightarrow \mathbb{R}$}
\int_X fdv = \int_X fgd\mu
\end{align}
