# Measures and functions

## Definitions

$(X,S)$ is a $\textbf{measurable space}$ if $S$ is a sigma algebra and $\bigcup S = X$

$(X,S,\mu)$ is a $\textbf{measure space}$ if $(X,S)$ is a measurable space and $\mu$ a measure defined on $S$

$E$ is a $\textbf{measurable set}$ if $E\in S$.

$\mu : \xi \rightarrow \mathbb{R}$ s.t $\xi$ is a class of sets of $X$ is a $\textbf{set function}$ 

$\mu : S \rightarrow \mathbb{R}$ is a $\textbf{measure}$ if it is non-negative, countably additive and $\mu(\emptyset) = 0$

$f : X \rightarrow \mathbb{R}$ is a $\textbf{measurable function}$ if $f^{-1}(M) = \{x \in X : f(x) \in M\}$ s.t $M \in 2^{\mathbb{R}}$ is measurable

$f_+$ is the $\textbf{positive part}$ of $f$ if:

\begin{align}
f_+(x) = 
\begin{cases}
f(x) \;&if \;f(x) > 0\\ 
0 \;&if \;f(x) \leq 0
\end{cases}
\end{align}

$f_-$ is the $\textbf{negative part}$ of $f$ if:

\begin{align}
f_-(x) = 
\begin{cases}
-f(x) \;&if \;f(x) < 0\\ 
0 \;&if \;f(x) \geq 0
\end{cases}
\end{align}



$\mathbb{1}_E : X \rightarrow \{0,1\}$ is a $\textbf{characteristic function}$ or $\textbf{indicator function}$ on the subset $E \subset X$ if :

\begin{align}
\mathbb{1}_E(x) = 
\begin{cases}
1 \;if \;x \in E\\ 
0 \;if \;x \notin E
\end{cases}
\end{align}

$f : X \rightarrow \mathbb{R}$ is a $\textbf{simple function}$ if for $a_i \in \mathbb{R}$ and $E_i \in S$ disjoints we have :

\begin{align} f(x) = \sum_{i=1}^na_i\mathbb{1}_{E_i}(x)\end{align}


## Properties of set functions

- Finitly additive : 
 $E_1 \dots E_n \in \xi \implies \mu(\sum_{i=1}^n E_i) = \sum_{i=1}^n \mu(E_i)$
 
- Countably additive : 
 $E_1 \dots \in \xi \implies \mu(\sum_{i=1}^{\infty} E_i) = \sum_{i=1}^{\infty} \mu(E_i) $

- Finite : 
 $E \in \xi \implies \left |\mu(E) \right| < \infty $

- Continuous from below

 > $E \in \xi$ 
 
 > $\forall i$, $E_i \in \xi$
 
 > $(E_n)_{n \in \mathbb{N}} \uparrow E$
 
 \begin{align}\lim_{n \to \infty}\mu(E_n) = \mu(E) \end{align}

- Continuous from above

 > $E \in \xi$ 
 
 > $\forall i$, $E_i \in \xi$
 
 > $(E_n)_{n \in \mathbb{N}} \downarrow E$
 
 > $\exists m$ s.t $|\mu(E_m)| < \infty$
 
 \begin{align}\lim_{n \to \infty}\mu(E_n) = \mu(E) \end{align}


## Extension of measures

> $P$ semiring 

> $R(P)$ ring generated by $P$

> $2^\Omega$ power set

> $S^*$ class of measurable sets

> $\sigma(P)$ sigma-ring generated by $P$

- Semiring $\rightarrow$ Ring

> $A \in R(P)$

> $E_i \in P$

> $\bigcup_{i=1}^{\infty} E_i \supset A$ 


\begin{align}
\mu_{R(P)}(A) = \sum_{i=1}^{\infty}\mu_P(E_i)
\end{align}

- Ring $\rightarrow$ Power set

> $A \in 2^\Omega$

> $E_i \in R(P)$

> $\bigcup_{i=1}^{\infty} E_i \supset A$ 


\begin{align}
\mu_{2^\Omega}(A) = \inf\{\sum_{i=1}^{\infty}\mu_{R(P)}(E_i)\}
\end{align}

- Power set $\rightarrow$ measurable sets

We retrict $\mu_{S^*} = \mu_{2^\Omega} \mid_{S^*}$

- measurable sets $\rightarrow$ sigma ring

We retrict $\mu_{\sigma(P)} = \mu_{2^\Omega} \mid_{\sigma(P)}$

## Lebesgue Measure

> $\Omega = \mathbb{R}$

> $P = \{(a,b]: a,b \in \mathbb{R}\}$ is a Semiring


Define a set function $\mu_{\sigma(R)}$ :
\begin{align}\mu_{\sigma(R)}((a,b]) = b - a\end{align}

$\mu$ has a unique extension $\mu_{\sigma(P)}$ to $B = \sigma(P)$, the class of Borel sets.

$\mu_{\sigma(P)}$ is the $\textbf{Lebesgue measure}$

If there exist a function $F : \mathbb{R} \rightarrow \mathbb{R}$ such that $F$ is non decreasing and right continuous at all points, we define a set function :

\begin{align}\mu_F((a,b]) = F(b) - F(a)\end{align}

$\mu_F$ is the $\textbf{Lebesgue-Stieltjes measure}$
